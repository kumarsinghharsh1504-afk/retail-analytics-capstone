```python
clean_base = raw_df.copy()

clean_base = clean_base.rename(columns={
    "Invoice": "invoice",
    "StockCode": "stock_code",
    "Description": "description",
    "Quantity": "quantity",
    "InvoiceDate": "invoice_date",
    "Price": "unit_price",
    "Customer ID": "customer_id",
    "Country": "country",
    "SourceYear": "source_year"
})

text_columns = [
    "invoice",
    "stock_code",
    "description",
    "country",
    "source_year"
]

for column in text_columns:
    clean_base[column] = (
        clean_base[column]
        .astype("string")
        .str.strip()
    )

clean_base["is_cancelled"] = (
    clean_base["invoice"]
    .str.startswith("C", na=False)
)

clean_base["is_negative_quantity"] = (
    clean_base["quantity"] < 0
)

clean_base["is_non_positive_price"] = (
    clean_base["unit_price"] <= 0
)

clean_base["is_missing_description"] = (
    clean_base["description"].isna()
    | clean_base["description"].eq("")
)

clean_base["is_missing_customer"] = (
    clean_base["customer_id"].isna()
)

print("Cleaning flags created")
display(clean_base.head())
```

```python
sales_mask = (
    clean_base["is_cancelled"].eq(False)
    & clean_base["quantity"].gt(0)
    & clean_base["unit_price"].gt(0)
    & clean_base["is_missing_description"].eq(False)
)

sales_df = (
    clean_base
    .loc[sales_mask]
    .copy()
)

excluded_df = (
    clean_base
    .loc[~sales_mask]
    .copy()
)

sales_df["customer_id"] = (
    sales_df["customer_id"]
    .astype("Int64")
)

sales_df["line_revenue"] = (
    sales_df["quantity"]
    * sales_df["unit_price"]
).round(2)

sales_df["invoice_year"] = (
    sales_df["invoice_date"].dt.year
)

sales_df["invoice_month"] = (
    sales_df["invoice_date"]
    .dt.to_period("M")
    .astype(str)
)

sales_df["invoice_hour"] = (
    sales_df["invoice_date"].dt.hour
)

print(
    "Clean sales rows:",
    f"{len(sales_df):,}"
)

print(
    "Excluded rows:",
    f"{len(excluded_df):,}"
)

display(sales_df.head())
```

```python
customer_df = (
    sales_df
    .dropna(subset=["customer_id"])
    .copy()
)

customer_df["customer_id"] = (
    customer_df["customer_id"]
    .astype("int64")
)

print(
    "Customer analysis rows:",
    f"{len(customer_df):,}"
)

print(
    "Unique customers:",
    f"{customer_df['customer_id'].nunique():,}"
)
```

```python
from pathlib import Path
import pandas as pd

project_path = Path(
    r"C:\Projects\Harsh Retail Analytics Capstone"
)

file_path = (
    project_path
    / "Raw Data"
    / "online_retail_II.xlsx"
)

clean_data_path = project_path / "Clean Data"

clean_data_path.mkdir(
    parents=True,
    exist_ok=True
)

if "raw_df" not in globals():
    excel_file = pd.ExcelFile(file_path)

    dataframes = []

    for sheet_name in excel_file.sheet_names:
        sheet_data = pd.read_excel(
            file_path,
            sheet_name=sheet_name
        )

        sheet_data["SourceYear"] = sheet_name
        dataframes.append(sheet_data)

    raw_df = pd.concat(
        dataframes,
        ignore_index=True
    )

print("Raw rows:", f"{len(raw_df):,}")

clean_base = raw_df.copy()

clean_base = clean_base.rename(columns={
    "Invoice": "invoice",
    "StockCode": "stock_code",
    "Description": "description",
    "Quantity": "quantity",
    "InvoiceDate": "invoice_date",
    "Price": "unit_price",
    "Customer ID": "customer_id",
    "Country": "country",
    "SourceYear": "source_year"
})

text_columns = [
    "invoice",
    "stock_code",
    "description",
    "country",
    "source_year"
]

for column in text_columns:
    clean_base[column] = (
        clean_base[column]
        .astype("string")
        .str.strip()
    )

rows_before_duplicates = len(clean_base)

clean_base = (
    clean_base
    .drop_duplicates()
    .copy()
)

duplicates_removed = (
    rows_before_duplicates
    - len(clean_base)
)

clean_base["is_cancelled"] = (
    clean_base["invoice"]
    .str.startswith("C", na=False)
)

clean_base["is_negative_quantity"] = (
    clean_base["quantity"] < 0
)

clean_base["is_non_positive_price"] = (
    clean_base["unit_price"] <= 0
)

clean_base["is_missing_description"] = (
    clean_base["description"].isna()
    | clean_base["description"].eq("")
)

clean_base["is_missing_customer"] = (
    clean_base["customer_id"].isna()
)

sales_mask = (
    clean_base["is_cancelled"].eq(False)
    & clean_base["quantity"].gt(0)
    & clean_base["unit_price"].gt(0)
    & clean_base["is_missing_description"].eq(False)
)

sales_df = (
    clean_base
    .loc[sales_mask]
    .copy()
)

excluded_df = (
    clean_base
    .loc[~sales_mask]
    .copy()
)

sales_df["customer_id"] = (
    sales_df["customer_id"]
    .astype("Int64")
)

sales_df["line_revenue"] = (
    sales_df["quantity"]
    * sales_df["unit_price"]
).round(2)

sales_df["invoice_year"] = (
    sales_df["invoice_date"].dt.year
)

sales_df["invoice_month"] = (
    sales_df["invoice_date"]
    .dt.to_period("M")
    .astype(str)
)

sales_df["invoice_hour"] = (
    sales_df["invoice_date"].dt.hour
)

customer_df = (
    sales_df
    .dropna(subset=["customer_id"])
    .copy()
)

customer_df["customer_id"] = (
    customer_df["customer_id"]
    .astype("int64")
)

cleaning_summary = pd.DataFrame({
    "metric": [
        "Raw rows",
        "Exact duplicates removed",
        "Rows after duplicate removal",
        "Clean sales rows",
        "Excluded rows",
        "Customer analysis rows",
        "Missing customer rows retained",
        "Unique clean invoices",
        "Unique customers",
        "Total positive revenue"
    ],
    "value": [
        len(raw_df),
        duplicates_removed,
        len(clean_base),
        len(sales_df),
        len(excluded_df),
        len(customer_df),
        sales_df["customer_id"].isna().sum(),
        sales_df["invoice"].nunique(),
        customer_df["customer_id"].nunique(),
        round(sales_df["line_revenue"].sum(), 2)
    ]
})

sales_df.to_csv(
    clean_data_path
    / "clean_sales_transactions.csv",
    index=False
)

customer_df.to_csv(
    clean_data_path
    / "customer_transactions.csv",
    index=False
)

excluded_df.to_csv(
    clean_data_path
    / "excluded_records_with_flags.csv",
    index=False
)

cleaning_summary.to_csv(
    clean_data_path
    / "cleaning_summary.csv",
    index=False
)

display(cleaning_summary)

print("\nSaved files:")

for file in clean_data_path.iterdir():
    print(file.name)
```

```python
from pathlib import Path
from getpass import getpass
import psycopg2

csv_path = Path(
    r"C:\Projects\Harsh Retail Analytics Capstone\Clean Data\clean_sales_transactions.csv"
)

postgres_password = getpass(
    "Enter your PostgreSQL password: "
)

conn = psycopg2.connect(
    host="localhost",
    port=5432,
    dbname="retail_capstone",
    user="postgres",
    password=postgres_password
)

print("Connected:", conn.closed == 0)
print("CSV exists:", csv_path.exists())
```

```python
copy_sql = """
COPY public.clean_sales_transactions (
    invoice,
    stock_code,
    description,
    quantity,
    invoice_date,
    unit_price,
    customer_id,
    country,
    source_year,
    is_cancelled,
    is_negative_quantity,
    is_non_positive_price,
    is_missing_description,
    is_missing_customer,
    line_revenue,
    invoice_year,
    invoice_month,
    invoice_hour
)
FROM STDIN
WITH (
    FORMAT CSV,
    HEADER TRUE,
    DELIMITER ',',
    QUOTE '"',
    ESCAPE '"',
    NULL '',
    ENCODING 'UTF8'
);
"""

try:
    with conn.cursor() as cursor:
        cursor.execute(
            "TRUNCATE TABLE public.clean_sales_transactions;"
        )

        with open(
            csv_path,
            mode="r",
            encoding="utf8",
            newline=""
        ) as csv_file:
            cursor.copy_expert(
                copy_sql,
                csv_file
            )

    conn.commit()
    print("Import completed successfully")

except Exception as error:
    conn.rollback()
    print("Import failed")
    print(error)
```

```python
with conn.cursor() as cursor:
    cursor.execute("""
        SELECT COUNT(*)
        FROM public.clean_sales_transactions;
    """)

    total_rows = cursor.fetchone()[0]

print(
    "PostgreSQL rows:",
    f"{total_rows:,}"
)
```

```python
with conn.cursor() as cursor:
    cursor.execute("""
        SELECT
            COUNT(*) AS total_rows,
            COUNT(DISTINCT invoice) AS unique_invoices,
            COUNT(DISTINCT customer_id) AS unique_customers,
            ROUND(SUM(line_revenue), 2) AS total_revenue,
            MIN(invoice_date) AS first_sale,
            MAX(invoice_date) AS last_sale
        FROM public.clean_sales_transactions;
    """)

    database_summary = cursor.fetchone()

print("Total rows:", f"{database_summary[0]:,}")
print("Unique invoices:", f"{database_summary[1]:,}")
print("Unique customers:", f"{database_summary[2]:,}")
print("Total revenue:", database_summary[3])
print("First sale:", database_summary[4])
print("Last sale:", database_summary[5])
```
